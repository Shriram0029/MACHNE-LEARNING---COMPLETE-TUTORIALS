**Imbalanced Dataset:**

a dataset with an unequal class distribution

```python
# importing libraries

import pandas as pd
import numpy as np
```

```python
# loading the dataset to pandas dataframe

credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
credit_card_data.isnull().sum()
```

```python
# distribution of two classes

credit_card_data['Class'].value_counts()
```

0 --> Legit transactions

1 --> Fraudelent transactions

This is highly imbalanced dataset

```python
# seperating the classes legit and fraudulent transactions

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

UNDER - SAMPLING IS USED TO HANDLE UNBALANCED DATA

Build a sample dataset containing similar distribution of Legit & Fraudulent transactions

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample.shape)
```

Concatenate the 2 dataframes legit_sample and fraud

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)

# axis = 0 will concatenate row wise
```
