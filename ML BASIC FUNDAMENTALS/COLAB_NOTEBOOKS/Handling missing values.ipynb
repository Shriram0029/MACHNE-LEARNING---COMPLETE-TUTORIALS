Methods to handle missing values

1.   Imputation
2.   Dropping

DATASET USED - Placement_Dataset.csv


Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# loading the dataset to pandas Dataframe

dataset = pd.read_csv("/content/Placement_Dataset.csv")
dataset.head()
```

```python
# to print the no. of rows and columns
dataset.shape
```

```python
# finding missing values in each column
dataset.isnull().sum()
```

Imputations method

Central Tendency:


1.   Mean
2.   Median
3.   Mode

```python
# analyse the distribution of data in salary as it contains missing values

fig, ax = plt.subplots(figsize=(20,8))
sns.distplot(dataset.salary)
```

Since the values are concentrated in one area and their are outliers , we will not use mean

Use MEDIAN OR MODE

Replace the missing values with median value

```python
dataset['salary'].fillna(dataset['salary'].median,inplace=True)
```

```python
# checking missing values
dataset.isnull().sum()
```

DROPPING VALUES (highly not recommended when dataset is small)

```python
salary_dataset = pd.read_csv("/content/Placement_Dataset.csv")
salary_dataset.isnull().sum()
```

```python
# drop the missing values
salary_dataset = salary_dataset.dropna(how='any')
```
