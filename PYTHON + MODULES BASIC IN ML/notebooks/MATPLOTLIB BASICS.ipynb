MATPLOTLIB

- used for making plots

```python
from matplotlib import pyplot as plt

# or

import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots

import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

PLOTTING THE DATA

```python
# sine wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# adding title with labels
plt.plot(x,y)
plt.xlabel('ANGLES')
plt.ylabel("SINE VALUES")
plt.title("SINE WAVE")
plt.show()
```

```python
# parabola

x = np.linspace(-10,10,20)
y = x**2

plt.plot(x,y)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("PARABOLA")
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g^')
plt.show()
```

```python
plt.plot(x,y,'bx')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x), 'r*')
plt.plot(x, np.cos(x), 'b+')
plt.show()
```

BAR GRAPH PLOTS

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]
ax.bar(languages, people)
plt.xlabel("LANGUAGES")
plt.ylabel("NO. OF PEOPLE")
plt.title("POPULAR LANGUAGES")
plt.show()
```

PIE CHART

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]
ax.pie(people, labels = languages, autopct = '%1.2f%%')
plt.show()
```

SCATTER PLOT

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y, color = 'r')
ax.scatter(x,z, color = 'b')
plt.show()
```

3D SCATTER PLOT

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c = z, cmap = 'Greens')
plt.show()
```
