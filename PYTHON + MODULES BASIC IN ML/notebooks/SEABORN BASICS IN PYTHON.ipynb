Importing the libraries


```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

NOTE : seaboen has some built in dataset

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x='total_bill', y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

SCATTER PLOT

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
```

```python
sns.scatterplot(x='petal_length', y='petal_length', hue='species', data=iris)
```

```python
# breast cancer dataset
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
temp  = pd.DataFrame(data.data, columns=data.feature_names)
temp.head()
```

DISTRIBUTION PLOT

```python
sns.displot(data=temp, x='mean radius', kde=True)
```

CORRELATION


*   positive correlation
*   negative correlation



HEAT MAPS


```python
correlation = temp.corr()

# constructing a heat map

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
