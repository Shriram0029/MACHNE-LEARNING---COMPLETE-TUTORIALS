MATHEMATICAL OPERATIONS IN NUMPY ARRAY

```python
a = np.random.randint(0,10,(3,3))
b = np.random.randint(10,20,(3,3))
print(a)
print(b)
```

```python
# add, subtract , multiply and divide elemnets in array
print(a+b)
print(a-b)
print(a*b)
print(a/b)
```

```python
print(np.add(a,b))
print(np.subtract(a,b))
print(np.multiply(a,b))
print(np.divide(a,b))
```

**ARRAY MANIPULATION**

```python
array = np.random.randint(0,10,(2,3))
print(array)
print(array.shape)
```

```python
# transpose of array matrix

trans = np.transpose(array)
print(trans)
print(trans.shape)

print()

trans2 = array.T
print(trans2)
print(trans2.shape)
```

```python
# reshaping a array

a = np.random.randint(0,10,(2,3))
print(a)
print(a.shape)

print()

b = a.reshape(3,2)
print(b)
print(b.shape)
```
